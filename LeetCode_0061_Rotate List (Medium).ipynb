### Problem Discription
Given the head of a linked list, rotate the list to the right by k places.

 

Example 1:  
  
Input: head = [1,2,3,4,5], k = 2  
Output: [4,5,1,2,3]  

Example 2:  
  
Input: head = [0,1,2], k = 4  
Output: [2,0,1]  

### Solution
The concept to solve this problem is quite simple, we just walk to the last kth node, and set the node's next value to None. Than we set the last node's next value to first node and we solve this problem.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if not head:
            return
        if not head.next:
            return head
        cur = tail = head
        l = 0
        while cur:
            l+=1
            tail = cur
            cur = cur.next

        k = k%l
        if k==0:
            return head
        
        mid = head
        for i in range(l-k-1):
            mid = mid.next
        res = mid.next
        mid.next = None
        tail.next = head
        
        return res
```
