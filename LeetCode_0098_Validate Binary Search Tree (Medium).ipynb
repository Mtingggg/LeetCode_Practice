### Problem Discription
Given the root of a binary tree, determine if it is a valid binary search tree (BST).

A valid BST is defined as follows:

The left subtree of a node contains only nodes with keys less than the node's key.
The right subtree of a node contains only nodes with keys greater than the node's key.
Both the left and right subtrees must also be binary search trees.

- Example 1:


Input: root = [2,1,3] <br>
Output: true <br>

- Example 2:


Input: root = [5,1,4,null,null,3,6] <br>
Output: false <br>
Explanation: The root node's value is 5 but its right child's value is 4.

### Solution
we can think this question in a really easy way, a valid binary tree means the node values from left to right(don't consider the depth) must have a ascending value
here is a easy example:

we can see the node values from left to right for not considering the depth is [1, 5, 3, 4, 6] and we can find that this wasn't in a ascending order, so this is not a valid binary tree

so what we will do is to go through every tree nodes in this order and compare if the current node value is larger than last node value

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def __init__(self):
        self.ans = True
        self.last_val = float('-inf')
    def isValidBST(self, node):
        if not (node.left or node.right):
            self.ans = self.ans&(node.val>self.last_val)
            self.last_val = node.val
            return self.ans  
        
        if node.left:
            self.isValidBST(node.left)
        self.ans = self.ans&(node.val>self.last_val)
        self.last_val = node.val
        if node.right:
            self.isValidBST(node.right)
            
        return self.ans
```
