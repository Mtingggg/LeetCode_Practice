## 1.暴力解

```python
s =  [1,8,6,2,5,4,8,3,7]

start = 0
end = 0
a = 0
temp_a = 0

for i in range(len(s)):
    for j in range(i, len(s)):
        
        temp_a = min(s[i], s[j])*(j-i)
        
        if temp_a>a:
            a = temp_a

a
```

## 2.挑小的移動
思路:兩個index，l表示最左r表示最右，只移動高度較低的idx向內移動，透過移動高度較低的值，期望找到一個高度較高的值得出更大的面積。

```python
s =  [1,8,6,2,5,4,8,3,7]

l = 0
r = len(s)-1
a = 0

while l!=r:
    a = max(a, min(s[l], s[r]) * (r-l))
    if s[l]>s[r]:
        r-=1
    else:
        l+=1
        
a        
```
